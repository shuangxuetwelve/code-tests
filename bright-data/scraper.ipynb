```python
import os
import requests

from dotenv import load_dotenv
load_dotenv()
```

```python
BRIGHT_DATA_API_KEY = os.getenv("BRIGHT_DATA_API_KEY")
if BRIGHT_DATA_API_KEY == None:
  raise Exception("Bright Data API key is not set")
print(f"Bright Data API key: *******************************")
```

```python
DATASET_ID_X_PROFILE = "gd_lwxmeb2u1cniijd7t4" # Dataset for scraping X profiles
```

```python
url_target = "https://x.com/elonmusk"

url = f"https://api.brightdata.com/datasets/v3/scrape?dataset_id={DATASET_ID_X_PROFILE}"

headers = {
  "Authorization": f"Bearer {BRIGHT_DATA_API_KEY}",
  "Content-Type": "application/json",
}

payload = { "input": [{"url": url_target, "max_number_of_posts": 100}] }

response = requests.request("POST", url, headers=headers, json=payload)

response.text
```
