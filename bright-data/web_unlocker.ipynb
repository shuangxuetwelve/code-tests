### Imports

````python
import os
import requests

from dotenv import load_dotenv
load_dotenv()

import html2text
````

### Load Environment Variables

````python
BRIGHT_DATA_API_KEY = os.getenv("BRIGHT_DATA_API_KEY")
if BRIGHT_DATA_API_KEY == None:
  raise Exception("Bright Data API key is not set")
print(f"Bright Data API key: *******************************")

BRIGHT_DATA_ZONE_WEB_UNLOCKER = os.getenv("BRIGHT_DATA_ZONE_WEB_UNLOCKER")
if BRIGHT_DATA_ZONE_WEB_UNLOCKER == None:
  raise Exception("Bright Data Web Unlocker zone is not set")
print(f"Bright Data Web Unlocker zone: *******************************")
````

````python
target_url = "https://techcrunch.com/2025/06/07/week-in-review-why-anthropic-cut-access-to-windsurf/"

url = "https://api.brightdata.com/request"

headers = {
    "Authorization": f"Bearer {BRIGHT_DATA_API_KEY}",
    "Content-Type": "application/json"
}

data = {
  "zone": BRIGHT_DATA_ZONE_WEB_UNLOCKER,
  "url": target_url,
  "format": "raw"
}

response = requests.post(url, headers=headers, json=data)
html = response.text

markdown = html2text.html2text(html)
print(markdown)
````
