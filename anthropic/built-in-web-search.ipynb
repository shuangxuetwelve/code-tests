### Imports and Initialization

````python
from pprint import pprint as pp

from dotenv import load_dotenv
load_dotenv()

import anthropic

client = anthropic.Anthropic()

def obj_to_dict(obj: any, indent=0):

  d = {}

  try:
    if type(obj) == list:
      return [obj_to_dict(item) for item in obj]
    else:
      properties = vars(obj)
      for k, v in properties.items():
        d[k] = obj_to_dict(v, indent + 1)
      return d
  except TypeError:
    return obj
````

### Send the Message

````python
# ~15s

messages = [
  {
    "role": "user",
    "content": "谁是鱼闪闪？"
  }
]

message = client.messages.create(
    model="claude-sonnet-4-20250514",
    system="You are a helpful assistant that can answer questions about the world.",
    messages=messages,
    max_tokens=1024, # max_tokens is the maxium tokens to generate, it is required
    tools=[{
        "type": "web_search_20250305",
        "name": "web_search"
    }]
)
````

````python
pp(obj_to_dict(message))
````
