# Test Tavily Map Feature

Tavily map会从一个网页开使，列出找到的链接。它的作用并且是获取site map，只是爬取链接。

## 资料

[API文档](https://docs.tavily.com/sdk/python/reference#tavily-map)

## Environments

Install Python packages:

```sh
pip install tavily-python python-dotenv
```

## Codes

### Imports and Initialization

````python
from pprint import pprint as pp

from dotenv import load_dotenv
load_dotenv()

from tavily import TavilyClient

client = TavilyClient()
````

### Basic Usage

````python
response = client.map(url="https://36kr.com/")

pp(response)
````
