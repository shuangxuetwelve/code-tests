````python
from pprint import pprint as pp

from dotenv import load_dotenv
load_dotenv()

from tavily import TavilyClient

client = TavilyClient()
````

````python
response = client.search("谁是鱼闪闪？")

pp(response)
````

### max_results

````python
# max_results defaults to 5
# 0 < max_results <= 20
response = client.search("AI", max_results=10)

pp(response)
````

### time_range

````python
# Available values for max_results:
# - day / d
# - week / w
# - month / m
# - year / y
#
# time_range defaults to not set
response = client.search("谁是鱼闪闪？", time_range="month")

pp(response)
````

### search_depth

将`search_depth`设置为`advanced`会让搜索到的source和每个搜索结果中的content都与query更加相关。

````python
# search_depth defaults to "basic"
response = client.search("谁是鱼闪闪？", search_depth="advanced")

pp(response)
````

### include_raw_content

无法保证一定会返回raw content，所以，对部分搜索结果，`raw_content` 仍然为`None`。

````python
# include_raw_content defaults to False
response = client.search("谁是鱼闪闪？", include_raw_content=True)

pp(response)
````

### include_images

````python
# include_images defaults to False
response = client.search("谁是鱼闪闪？", include_images=True)

pp(response)
````

### include_image_descriptions

`include_image_descriptions`设置为`True`时，也会返回相应的image URL。

````python
# include_image_descriptions defaults to False
response = client.search("谁是鱼闪闪？", include_image_descriptions=True)

pp(response)
````

### include_domains

设置了`include_domains`会让搜索结果只来源于这些domain中。

````python
# include_domains defaults to []
response = client.search("谁是鱼闪闪？", include_domains=["youtube.com"])

pp(response)
````

### exclude_domains

设置了`exclude_domains`之后，搜索结果就不会来源于这些domain了。

````python
# exclude_domains defaults to []
response = client.search("谁是鱼闪闪？", exclude_domains=["baidu.com"])

pp(response)
````
