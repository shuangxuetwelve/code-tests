## Codes

### Imports and Initialization

````python
from pprint import pprint as pp

from dotenv import load_dotenv
load_dotenv()

from tavily import TavilyClient

client = TavilyClient()
````

### Basic Usage

````python
response = client.extract(urls=["https://36kr.com/information/AI/"])

pp(response)
````

### extract_depth

* basic: 只能提取到静态网页，而且不能提取到table和embeded content
* advanced: 可以提取到动态网页，还可以提取到table和embeded content

Cost
* basic: 1 credit / 5 extractions
* advanced: 2 credits / 5 extractions

````python
# extract_depth defaults to "basic"
response = client.extract(urls=["https://www.douyin.com/jingxuan"], extract_depth="advanced")

pp(response)
````
