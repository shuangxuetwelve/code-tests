# Test Tavily Crawl Feature

Tavily crawl会从一个网页开始，不断爬取更多的网页。

## 资料

[API文档](https://docs.tavily.com/sdk/python/reference#tavily-crawl)

## Environments

Install Python packages:

```sh
pip install tavily-python python-dotenv
```

## Codes

### Imports and Initialization

````python
from pprint import pprint as pp

from dotenv import load_dotenv
load_dotenv()

from tavily import TavilyClient

client = TavilyClient()
````

### Basic Usage

````python
response = client.crawl(url="https://36kr.com/")

print("Number of pages crawled:", len(response['results']))
pp(response)
````

### Instructions

`instructions`参数会控制爬取过程中只爬取与`instructions`参数有关的网页。但是我没有看到明确的效果。

````python
response = client.crawl(url="https://36kr.com/", instructions="AI news", timeout=100)

print("Number of pages crawled:", len(response['results']))
pp(response)
````
