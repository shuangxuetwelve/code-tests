```python
import os
import requests
from pprint import pp

from dotenv import load_dotenv
load_dotenv()
```

```python
JINA_API_KEY = os.getenv("JINA_API_KEY")
if JINA_API_KEY == None:
  raise Exception("Jina AI API key is not set")
print(f"Jina AI API key is {JINA_API_KEY}")
```

```python
url_target = "https://techcrunch.com/latest/"
```

### Default Usage

```python
url = f"https://r.jina.ai/{url_target}"

headers = {
    "Authorization": f"Bearer {JINA_API_KEY}",
}

response = requests.get(url, headers=headers)

print(f"Response status code: {response.status_code}")
response.encoding = "utf-8"
pp(response.text)
```

### JSON Response

```python
url = f"https://r.jina.ai/{url_target}"

headers = {
    "Authorization": f"Bearer {JINA_API_KEY}",
    "Accept": "application/json"
}

response = requests.get(url, headers=headers)

print(f"Response status code: {response.status_code}")
response.encoding = "utf-8"
pp(response.json())
```

### JSON Response & Markdown Content

```python
url = f"https://r.jina.ai/{url_target}"

headers = {
    "Authorization": f"Bearer {JINA_API_KEY}",
    "Accept": "application/json",
    "X-Return-Format": "markdown"
}

response = requests.get(url, headers=headers)

print(f"Response status code: {response.status_code}")
response.encoding = "utf-8"
pp(response.json())
```

### JSON Response & HTML content

```python
url = f"https://r.jina.ai/{url_target}"

headers = {
    "Authorization": f"Bearer {JINA_API_KEY}",
    "Accept": "application/json",
    "X-Return-Format": "html"
}

response = requests.get(url, headers=headers)

print(f"Response status code: {response.status_code}")
response.encoding = "utf-8"
pp(response.json())
```

### Gather Links at the End

```python
url = f"https://r.jina.ai/{url_target}"

headers = {
    "Authorization": f"Bearer {JINA_API_KEY}",
    "Accept": "application/json",
    "X-Return-Format": "markdown",
    "X-With-Links-Summary": "trup"
}

response = requests.get(url, headers=headers)

print(f"Response status code: {response.status_code}")
response.encoding = "utf-8"
pp(response.json())
```
